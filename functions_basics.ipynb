# Q1. Addition Function

## Problem Statement
Create a function that takes two numbers and returns their sum.

## Sample Input
```python
10, 20
```

## Sample Output
```python
30
```


````python
def sum():
    num1 = 10
    num2 = 20
    return num1 + num2
print(sum())
````


# Q7. Average Function

## Problem Statement
Create a function that takes marks of 3 subjects and returns the average.

## Sample Input
```python
70, 80, 90
```

## Sample Output
```python
80.0
```


````python
def average():
    a = 70
    b = 80
    c = 90
    return (a + b + c) / 3
print(average())
````


# Q8. Maximum Number Function

## Problem Statement
Create a function that takes two numbers and returns the greater number.

## Sample Input
```python
25, 40
```

## Sample Output
```python
40
```


````python
def greater():
    a = 25
    b = 40
    if a > b:
        return a
    else:
        return b
print(greater())
````


# Q10. Full Name Function

## Problem Statement
Create a function that takes first name and last name and returns the full name.

## Sample Input
```python
Rahul, Sharma
```

## Sample Output
```python
Rahul Sharma
```


````python
def name():
    name1 = "Rahul"
    name2 = "Sharma"
    return name1 + " " + name2
print(name())
````
