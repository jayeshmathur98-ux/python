Q2. Write a Python program to count total vowels in a string.

Q5. Write a Python program to count total uppercase and lowercase letters in a string.
   
    INPUT- PyThOn
    
    OUTPUT - Uppercase letters: 3
             Lowercase letters: 3

```python
a = "PyTHon"
upper = 0
lower = 0 
for i in a:
    if i .isupper():
        upper += 1
    elif i .islower():
        lower += 1
print("Uppercase letters:",upper)
print("lowercase letters:",lower)           
```
