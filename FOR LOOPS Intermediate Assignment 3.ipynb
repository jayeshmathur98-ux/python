Q4. Write a Python program to find factorial of a number.

    INPUT - Enter a number: 5
    
    OUTPUT - Factorial = 120

```python
n = int(input("enter a number :-"))
fact = 1
for i in range(1,n+1):
    fact = fact*i
print(f"factorial = ",fact)    
```

Q8. Write a Python program to count vowels and consonants in a string.

    INPUT - Enter a string: Python

    OUTPUT - Vowels = 1
             Consonants = 5

```python
s = input("enter a string:")
vowels = 0
consonants = 0
for ch in s:
    if ch.isalpha():
        if ch.lower() in "aeiou":
            vowels += 1
        else:
            consonants += 1
print("vowels = ",vowels)
print("consonants =",consonants)                
```

Q10. Write a Python program to find sum of digits of a number.

     INPUT - Enter a number: 123
     
     OUTPUT - 6

```python
num = int(input("enter a number:"))
sum = 0
for digit in str(num):
    sum = sum + int(digit)
print(sum)    
```
