Q4. Write a Python program to take a string from the user and perform slicing.
    Print:
    - First 5 characters
    - Characters from index 2 to 7
    - Last 4 characters
    
    INPUT - Programming
    
    OUTPUT - 
    
    First 5 characters: Progr
    Characters from index 2 to 7: ogramm
    Last 4 characters: ming
    

```python
text = input("enter a string :")
print("first 5 characters:", text[:5])
print("characters from index 2 to 7:", text[2:8])
print("last 4 characters:", text[-4:])
```

Q10. Write a Python program to check whether a word exists in the string or not.
      
      INPUT - Enter a sentence: Python is easy

      OUTPUT - True
